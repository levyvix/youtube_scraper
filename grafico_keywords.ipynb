```python
import pandas as pd
```

```python
df = pd.read_csv("videos_sorted.csv")
```

```python
df = df.drop_duplicates(subset='title')
```

```python
# get keywords

keywords = df.loc[df['keywords'].apply(lambda x: "['']" not in str(x)), 'keywords']
```

```python
dicionario = {}

for i in keywords:
    for j in i[1:-1].split(','):
        if j.strip() not in dicionario:
            dicionario[j.strip()] = 1
        else:
            dicionario[j.strip()] += 1
```

```python
# clean the keys

for i in list(dicionario.keys()):
    if i[0] == "'" and i[-1] == "'":
        dicionario[i[1:-1]] = dicionario.pop(i)

```

```python
# bar braph

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
plt.bar(dicionario.keys(), dicionario.values())
plt.xticks(rotation=90)
plt.show()
```
